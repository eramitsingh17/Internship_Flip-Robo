# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url_imdb = requests.get('https://www.imdb.com/chart/top')
url_imdb
```

```python
soup_imdb = BeautifulSoup(url_imdb.content)
soup_imdb
```

# Scrapping Movies Name

```python
Movie_Name = [] #Empty list to store the data
for i in soup_imdb.find_all('td', class_="titleColumn"):
    for i1 in i.find_all('a'):
        Movie_Name.append(i1.text)
    
Movies = Movie_Name[:100]
Movies
```

```python
IMDB_Rating = [] #Empty list to store the data
for i in soup_imdb.find_all('td', class_="ratingColumn imdbRating"):
    
    IMDB_Rating.append(i.text.replace('\n',''))
Rating = IMDB_Rating[:100]
Rating
```

```python
Movie_Release_Date = [] #Empty list to store the data
for i in soup_imdb.find_all('span', class_="secondaryInfo"):
    #x= i.text.replace("(","")
    Movie_Release_Date.append(i.text.replace("(","").replace(")",""))
       
Movie_Release_Date[:100]
Release_Date = Movie_Release_Date[:100]
Release_Date
```

```python
print(len(Movies),len(Rating),len(Release_Date))
```

```python
#Making dataframe
import pandas as pd
df = pd.DataFrame({'Movies-Names':Movies,'Movie-Ratings':Rating,'Movie-Release':Release_Date})
df
```

# 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year ofrelease) and make data frame.

```python
url_Hindi = requests.get('https://www.imdb.com/list/ls009997493/')
url_Hindi
```

```python
Hindi_Movie = [] #Empty list to store the data
for i in soup_Hindi.find_all('h3', class_="lister-item-header"):
    for i1 in i.find_all('a'):
        Hindi_Movie.append(i1.text.replace('\n',''))
HindiMovies = Hindi_Movie[:100]
HindiMovies
```

# Scrapping Movie Ratings

```python
HindiMovie_Rating = [] #Empty list to store the data
for i in soup_Hindi.find_all('div', class_="ipl-rating-star small"):
    
    HindiMovie_Rating.append(i.text.replace('\n',''))
MovieRating = HindiMovie_Rating[:100]
MovieRating
```

# Scrapping year of release

```python
Year_Release = [] #Empty list to store the data
for i in soup_Hindi.find_all('span', class_="lister-item-year text-muted unbold"):
    
    Year_Release.append(i.text.replace("(","").replace(")","").replace('I ',''))
Year_Release[:100]
Movie_Release_Year = Year_Release[:100]
Movie_Release_Year
```

```python
#Making dataframe
import pandas as pd
df = pd.DataFrame({'Hindi Movies Names':HindiMovies,'Hindi Movie Ratings':MovieRating,'Hindi Movie Release':Movie_Release_Year})
df
```

# 8) Write a python program to scrape mentioned details from dineout.co.in :


i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
url_dine = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
url_dine
```

```python
soup_dine = BeautifulSoup(url_dine.content)
soup_dine
```

# Scrapping Restaurant Name

```python
Rest_Name = []
for i in soup_dine.find_all('div', class_='restnt-info cursor'):
    for i1 in i.find_all('a'):
        Rest_Name.append(i1.text)
Rest_Names = Rest_Name[:13]
Rest_Names
```

# Scrapping Cuisine

```python
Cuisine = []
for i in soup_dine.find_all('span', class_='double-line-ellipsis'):
    for i1 in i.find_all('a'):
        Cuisine.append(i1.text)
Cuisines = Cuisine[:13]
Cuisines
```

# Scrapping Location

```python
Location = []
for i in soup_dine.find_all('div', class_='restnt-loc ellipsis'):
    Location.append(i.text)
Location
```

# Scrapping Ratings

```python
Ratings = []
for i in soup_dine.find_all('div', class_='restnt-rating rating-4'):
    Ratings.append(i.text)
Ratings
```

# Scrapping Image URL

```python
Images_url = []
for i in soup_dine.find_all('img', class_='no-img'):
    Images_url.append(i['data-src'])
Images_url
```

```python
print(len(Rest_Names),len(Cuisines),len(Location),len(Ratings),len(Images_url))
```

```python
df = pd.DataFrame({'Rest-Names':Rest_Names,'Cuis': Cuisines,'Loc': Location,'Rates': Ratings,'Images-url': Images_url})
df
```
