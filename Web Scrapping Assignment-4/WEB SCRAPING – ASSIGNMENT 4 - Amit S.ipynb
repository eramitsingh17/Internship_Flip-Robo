# 5. Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details:
- A) Repository title
- B) Repository description
- C) Contributors count
- D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
def trending_rep():
    """Creating function that searches details of GITHUB trendings repository"""
    
    template = 'https://github.com/' #URL template for accessing the website.
    driver = webdriver.Chrome(r'E:/Flip Robo Internship/chromedriver_win32/chromedriver.exe') #Calling the Web Driver.
    driver.get(template) #Opening with the URL template.
    driver.maximize_window()#Maximize the Window.
    time.sleep(5)#Making the Function wait for 3sec so webpage will open.
    
    #Searching the button trending and calling it in the web Driver.
    button = driver.find_element_by_xpath("//ul [@class= 'list-style-none mb-3']/li [@class = 'edge-item-fix']/a [@data-ga-click = '(Logged out) Header, go to Trending']")
    driver.get(button.get_attribute('href'))
    time.sleep(1) #Making the Function wait for 1sec so webpage will open.
    
    href = [] #Declaring the list href to scrap all the href details of trending repository.
    for i in driver.find_elements_by_xpath("//h1 [@class = 'h3 lh-condensed']/a"):
        href.append(i.get_attribute('href'))
        
                    
    driver.close() #Closing the Driver post collecting the Required Details.
    Repository_title = [] #Creating Repository_title list to collect the Repository_title details.
    Repository_description = [] #Creating Repository_description list to collect the Repository_description details.
    Contributors_count = [] #Creating Contributors_count list to collect the Contributors_count details.
    Language_used = [] #Creating Language_used list to collect the Language_used details.
    
    for i in href:#Iterating all the href and scraping the required details.
        driver = webdriver.Chrome(r'E:/Flip Robo Internship/chromedriver_win32/chromedriver.exe')#Calling the Web Driver.
        driver.get(i)#Opening with the iterating href details.
        driver.maximize_window()#Maximize the Window.
        time.sleep(2) #Making the Function wait for 1sec so webpage will open.
        
        #Scraping the Repository_title
        tit = driver.find_element_by_xpath("//h1 [@class = ' d-flex flex-wrap flex-items-center break-word f3 text-normal']")
        Repository_title.append(tit.text.replace('\n', ' '))
        
        #Scraping the Repository_description.
        try:
            des = driver.find_element_by_xpath("//div [@class = 'BorderGrid-cell']/p [@class = 'f4 mt-3']")
            Repository_description.append(des.text)
        except:
            Repository_description.append("-")
        
        #Scraping the Contributors_count.
        count = [] #Creating a Dummy Variable to collect the appropriate Contributors_count details.
        try:
            for i in driver.find_elements_by_xpath("//h2 [@class = 'h4 mb-3']//span [@class = 'Counter ']"):
                count.append(i.get_attribute('title'))
            Contributors_count.append(count[-1])
        except:
            Contributors_count.append('1')
                
        #Scraping the Language_used.
        lang = [] #Creating a Dummy Variable to collect the appropriate Language_used details.
        for i in driver.find_elements_by_xpath("//ul [@class = 'list-style-none']/li [@class = 'd-inline']//span"):
            lang.append(i.text)
        if (" ".join(lang)) == '':
            Language_used.append('-')
        else:
            Language_used.append(" ".join(lang))
        
        driver.close() #Quiting the Driver post scraping the Details.
    
    
    #Creating the DataFrame with the collected details.
    table = pd.DataFrame({"Repository title" : Repository_title,
                          "Repository description" : Repository_description,
                          "Contributors count" : Contributors_count,
                          "Language used" : Language_used})
    
    return table #returing the DataFrame Created.
```

```python
rep = trending_rep() #Calling the Function and storing it in the variable.
rep
```

# 6. Scrape the details of top 100 songs on billboard.com. Url = https://www.billboard.com/ You have to find the following details:
- A) Song name
- B) Artist name
- C) Last week rank
- D) Peak rank
- E) Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
#Activating the chrome browser
driver = webdriver.Chrome(r'E:/Flip Robo Internship/chromedriver_win32/chromedriver.exe')
time.sleep(3)

# Opening the homepage of top 100 songs on billboard.com.
url = "https://www.billboard.com/"
driver.get(url)

time.sleep(3)
search_button=driver.find_element_by_xpath('//div[@class="header__sticky__child"]/nav[1]/ul[1]/li[3]/a[1]')      #locating of search_button from xpath 
search_button.click()

#creating the empty list
hot_song = {}
hot_song['Song_name']=[]
hot_song['Artist_name']=[]
hot_song['Last_week_rank']=[]
hot_song['Peak_rank']=[]
hot_song['Weeks_on_board']=[]

time.sleep(3)
for i in range(0,1):
    # Song name
    song = driver.find_elements_by_xpath('//div[@class="chart-list__wrapper"]/div[1]/ol[1]/li/button[1]/span[2]/span[1]'.format(i))
    for s in song:
        hot_song['Song_name'].append(s.text)  
        
    # Artist name
    artist = driver.find_elements_by_xpath('//div[@class="chart-list__wrapper"]/div[1]/ol[1]/li/button[1]/span[2]/span[2]'.format(i))
    for a in artist:
        hot_song['Artist_name'].append(a.text)
    
    # Last week rank
    lws = driver.find_elements_by_xpath('//div[@class="chart-list__wrapper"]/div[1]/ol[1]/li/button[1]/div[1]/div[1]'.format(i))
    for l in lws:
        hot_song['Last_week_rank'].append(l.text)
    
    # Peak rank
    pr = driver.find_elements_by_xpath('//div[@class="chart-list__wrapper"]/div[1]/ol[1]/li/button[1]/div[1]/div[2]'.format(i))
    for p in pr:
        hot_song['Peak_rank'].append(p.text)

    # Weeks on board
    wob = driver.find_elements_by_xpath('//div[@class="chart-list__wrapper"]/div[1]/ol[1]/li/button[1]/div[1]/div[3]'.format(i))
    for w in wob:
        hot_song['Weeks_on_board'].append(w.text)
        
time.sleep(3)
# Make dataframe
hot_song_df = pd.DataFrame(hot_song)
hot_song_df.head()
```

# 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details:
- A) Dataset name
- B) Data type
- C) Task
- D) Attribute type
- E) No of instances
- F) No of attribute
- G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome(r'E:/Flip Robo Internship/chromedriver_win32/chromedriver.exe')
time.sleep(3)

# Opening the homepage of Datasets from UCI machine learning repositories.
url = "https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(3)

search_button = driver.find_element_by_xpath('//span[@class="normal"]/b[1]/a[1]')          #locating of search_button from xpath
search_button.click()

#creating the empty list
UCI_machine_learning_repositories = {}
UCI_machine_learning_repositories['Dataset name']=[]
UCI_machine_learning_repositories['Data type']=[]
UCI_machine_learning_repositories['Default Task']=[]
UCI_machine_learning_repositories['Attribute type']=[]
UCI_machine_learning_repositories['No of instances']=[]
UCI_machine_learning_repositories['No of attribute']=[]
UCI_machine_learning_repositories['Year']=[]

time.sleep(3)
for i in range(0,1):
    try:
        # Dataset name
        name = driver.find_elements_by_xpath('//p[@class="normal"]/b[1]/a[1]'.format(i))
        for n in name:
            UCI_machine_learning_repositories['Dataset name'].append(n.text)
        
        # Data type
        type = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[2]/p[1]'.format(i))
        for t in type:
            UCI_machine_learning_repositories['Data type'].append(t.text) 
            
        # Task
        task = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[3]/p[1]'.format(i))
        for t in task:
            UCI_machine_learning_repositories['Default Task'].append(t.text)
        
        # Attribute type
        attribute_type = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[4]/p[1]'.format(i))
        for at in attribute_type:
            UCI_machine_learning_repositories['Attribute type'].append(at.text)
            
        # No of instances
        instances = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[5]/p[1]'.format(i))
        for i in instances:
            UCI_machine_learning_repositories['No of instances'].append(i.text)
            
        # No of attribute
        no_of_attribute = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[6]/p[1]'.format(i))
        for noa in no_of_attribute:
            UCI_machine_learning_repositories['No of attribute'].append(noa.text)
            
        # Year
        year = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[7]/p[1]'.format(i))
        for y in year:
            UCI_machine_learning_repositories['Year'].append(y.text)
            
    except NoSuchElementException:
        UCI_machine_learning_repositories['Dataset name'].append('-')
        UCI_machine_learning_repositories['Data type'].append('-')
        UCI_machine_learning_repositories['Default Task'].append('-')
        UCI_machine_learning_repositories['Attribute type'].append('-')
        UCI_machine_learning_repositories['No of instances'].append('-')
        UCI_machine_learning_repositories['No of attribute'].append('-') 
        UCI_machine_learning_repositories['Year'].append('-')  
        
time.sleep(3)
# Make dataframe       
machine_learning=pd.DataFrame({})
machine_learning['Dataset name']=UCI_machine_learning_repositories['Dataset name'][:500]
machine_learning['Data type']=UCI_machine_learning_repositories['Data type'][:500]
machine_learning['Default Task']=UCI_machine_learning_repositories['Default Task'][:500]
machine_learning['Attribute typ']=UCI_machine_learning_repositories['Attribute type'][:500]
machine_learning['No of instances']=UCI_machine_learning_repositories['No of instances'][:500]
machine_learning['No of attribute']=UCI_machine_learning_repositories['No of attribute'][:500]
machine_learning['Year']=UCI_machine_learning_repositories['Year'][:500]
machine_learning.head()
```
